```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the California Housing dataset
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

print("First 5 rows of the dataset:")
print(df.head())

print("\nDataset Info:")
df.info()

print("\nDescriptive Statistics:")
print(df.describe())

# Plot a histogram of a numerical column, e.g., 'median_house_value'
plt.figure(figsize=(10, 6))
df['median_house_value'].hist(bins=50)
plt.title('Distribution of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generate some dummy data for Experience and Salary
np.random.seed(42) # for reproducibility
experience = np.random.uniform(1, 20, 100) # Experience in years
salary = 30000 + (experience * 3000) + np.random.normal(0, 5000, 100) # Base salary + experience effect + noise

salary_df = pd.DataFrame({
    'Experience': experience,
    'Salary': salary
})

print("Sample Salary Data:")
display(salary_df.head())
```

```python
# Create a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(salary_df['Experience'], salary_df['Salary'], alpha=0.7)
plt.title('Salary vs. Experience')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary (USD)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# --------------------------------
# Creating Sensor Dataset
# --------------------------------
data = {
    'Temperature': [30, 32, 31, 35, 36, 34, 37],
    'Humidity': [60, 65, 63, 70, 72, 68, 75]
}
# --------------------------------
# Creating DataFrame
# --------------------------------
df = pd.DataFrame(data)
print("Sensor Data")
print(df)
# --------------------------------
# Statistical Analysis
# --------------------------------
print("\nStatistical Summary")
print(df.describe())
# --------------------------------
# Finding Maximum and Minimum Values
# --------------------------------
print("\nMaximum Temperature:",
      df['Temperature'].max())
print("Minimum Temperature:",
      df['Temperature'].min())
print("Average Temperature:",
      round(df['Temperature'].mean(), 2))
print("\nMaximum Humidity:",
      df['Humidity'].max())
print("Minimum Humidity:",
      df['Humidity'].min())
print("Average Humidity:",
      round(df['Humidity'].mean(), 2))
# --------------------------------
# Visualization using Seaborn
# --------------------------------
sns.lineplot(data=df)
plt.title("Sensor Data Analysis")
plt.xlabel("Reading Number")
plt.ylabel("Sensor Value")
plt.show()
```
